# 4. Produtos Mais e Menos Vendidos

```python
import pandas as pd


def produtos_mais_menos_vendidos(df, nome_loja):
    contagem_produtos = df['Produto'].value_counts()

    mais_vendido = contagem_produtos.idxmax()
    menos_vendido = contagem_produtos.idxmin()

    print(f"\nüìç {nome_loja}")
    print(f"Produto MAIS vendido: {mais_vendido} - {contagem_produtos.max()} vendas")
    print(f"Produto MENOS vendido: {menos_vendido} - {contagem_produtos.min()} vendas")


produtos_mais_menos_vendidos(loja, "Loja 1")
produtos_mais_menos_vendidos(loja2, "Loja 2")
produtos_mais_menos_vendidos(loja3, "Loja 3")
produtos_mais_menos_vendidos(loja4, "Loja 4")


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Fun√ß√£o para gerar gr√°fico de linhas por loja
def grafico_produtos_linha(df, nome_loja):
    contagem = df['Produto'].value_counts().sort_values(ascending=False)

    plt.figure()
    plt.plot(contagem.index, contagem.values, marker='o')
    plt.title(f"Produtos mais e menos vendidos - {nome_loja}")
    plt.xlabel("Produto")
    plt.ylabel("Quantidade de vendas")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

# Aplicando para cada loja
grafico_produtos_linha(loja, "Loja 1")
grafico_produtos_linha(loja2, "Loja 2")
grafico_produtos_linha(loja3, "Loja 3")
grafico_produtos_linha(loja4, "Loja 4")
```
